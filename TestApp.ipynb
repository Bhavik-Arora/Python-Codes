```python
# My application
class BhavikServices:
    def __init__(self):
        self.bookings = []

    def book_service(self, client_name, service_type, date):
        booking = {
            'client_name': client_name,
            'service_type': service_type,
            'date': date
        }
        self.bookings.append(booking)
        print(f"Booking successful! {client_name} booked {service_type} on {date}")

    def view_bookings(self):
        if not self.bookings:
            print("No bookings yet.")
        else:
            print("Upcoming Bookings:")
            for booking in self.bookings:
                print(f"{booking['client_name']} - {booking['service_type']} on {booking['date']}")

    def cancel_booking(self, client_name, service_type, date):
        for booking in self.bookings:
            if booking['client_name'] == client_name and booking['service_type'] == service_type and booking['date'] == date:
                self.bookings.remove(booking)
                print(f"Booking for {client_name} ({service_type} on {date}) cancelled.")
                return
        print("Sorry, booking not found.")

    def check_availability(self, date):
        for booking in self.bookings:
            if booking['date'] == date:
                print(f"Sorry, Bhavik is not available on this date. Date:- {date}.")
                return
        print(f"Bhavik is available on {date}.")

def main():
    bhavik_services = BhavikServices()

    while True:
        print("\nWelcome to Bhavik's Services Test App v1.00")
        print("1. Book a service")
        print("2. View bookings")
        print("3. Cancel booking")
        print("4. Check availability")
        print("5. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            client_name = input("Enter your name: ")
            service_type = input("Enter the service you want: ")
            date = input("Enter the date (YYYY-MM-DD): ")
            bhavik_services.book_service(client_name, service_type, date)

        elif choice == '2':
            bhavik_services.view_bookings()

        elif choice == '3':
            client_name = input("Enter your name: ")
            service_type = input("Enter the service you want to cancel: ")
            date = input("Enter the date of booking (YYYY-MM-DD): ")
            bhavik_services.cancel_booking(client_name, service_type, date)

        elif choice == '4':
            date = input("Enter the date to check availability (YYYY-MM-DD): ")
            bhavik_services.check_availability(date)

        elif choice == '5':
            print("Exiting Bhavik's Services App. Goodbye!")
            break

        else:
            print("Invalid choice. Please enter a number from 1 to 5.")

if __name__ == "__main__":
    main()
```
